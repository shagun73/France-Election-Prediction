```python
import json
import pandas as pd
import re
from nltk.tokenize import word_tokenize

textl = []
idl = []
timel = []
platforml = []
useridl = []
locationl = []
candidatel = []
partyl = []

with open('demo.json', 'r') as f:
    for line in f:
        tweet = json.loads(line)
        textl.append(tweet['text'])
        idl.append(tweet['id'])
        timel.append(tweet['created_at'])
        platforml.append(''.join(re.findall(r">(.+?)<", tweet['source'])))
        useridl.append(tweet['user']['id'])
        if tweet['coordinates']:
           locationl.append(tweet['coordinates']['coordinates'])
        else:
           locationl.append(tweet['coordinates'])

data = {'id': idl, 'text': textl, 'created_at': timel,'platform':platforml,'user_id':useridl,
        'location':locationl}
df = pd.DataFrame(data)
```
